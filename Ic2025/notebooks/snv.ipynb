```python
import pandas as pd
import os

#Fazendo o caminho para o arquivo em csv
dir = (r"C:\Users\LucAA\Documents\Projetos\Ic2025\data_processed")

arquivo_csv = "casca_maca.csv"
path =os.path.join(dir,arquivo_csv)

df = pd.read_csv(path)

print(df.shape)
print(df.head())
```

```python
df_snv = df.apply(lambda linha: (linha - linha.mean())/  linha.std(), axis = 1)
#muda apenas no data frame
df_snv
```

```python
arquivo_snv = "casca_maca_snv.csv"

path_snv = os.path.join(dir, arquivo_snv)

df_snv.to_csv(path_snv,index =False)
```
