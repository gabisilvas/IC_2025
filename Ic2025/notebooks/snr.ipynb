```python
import pandas as pd
import numpy as np
import os

# Caminho do arquivo com SNV aplicado
dir_path = r"C:\Users\LucAA\Documents\Projetos\Ic2025\data_processed"
file_name = "casca_maca_ema.csv"
path = os.path.join(dir_path, file_name)
```

```python
df = pd.read_csv(path)

df_numeric = df.select_dtypes(include='number')
```

```python
snr = df_numeric.mean() / df_numeric.std()

limite_snr = 3 #pesquisar

# Mantém somente as bandas com SNR >= limite
bandas_filtradas = snr[snr >= limite_snr].index
df_filtrado = df_numeric[bandas_filtradas]

# Juntando de volta com as colunas não numéricas, se existirem
df_result = pd.concat([df.drop(columns=df_numeric.columns), df_filtrado], axis=1)
df_result.head()
```

```python
output_file = "casca_maca_snr.csv"
output_path = os.path.join(dir_path, output_file)
df_result.to_csv(output_path, index=False)
```
